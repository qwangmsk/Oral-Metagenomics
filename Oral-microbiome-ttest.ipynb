### Perform Independent Two Sample t-Test
An independent two sample t-test is used to determine if two population means are equal (Reference: https://www.statology.org/pandas-t-test/).

```python
from scipy.stats import ttest_ind
```

```python
group1 = df.loc['G1','Metabolism; Carbohydrate metabolism']
group2 = df.loc['G2','Metabolism; Carbohydrate metabolism']
```

```python
ttest_ind(group1, group2) #, equal_var=False)
```

```python
# Define a function 
```

```python
def my_ttest(gene):
    group1 = df.loc['G1',gene]
    group2 = df.loc['G2',gene]
    result = ttest_ind(group1, group2)
    return result
```

```python
# Test the function
# my_ttest('Metabolism').pvalue
```

```python
# Test each pathway and store the result into 
column_headers = df.columns.values.tolist()
results = pd.Series(dict((i, my_ttest(i).pvalue) for i in column_headers))
```
